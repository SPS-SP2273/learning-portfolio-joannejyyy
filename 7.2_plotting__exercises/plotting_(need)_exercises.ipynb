```python
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= x_err, yerr= y_err,
             color= 'red', linestyle= '--',
             marker= 'yellow', markerfacecolor= 'yellow',
             ecolor= 'blue',capsize=5,
             label= 'yes')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

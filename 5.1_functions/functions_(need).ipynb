```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    else:
        return f'Hello {name}!'
```

```python
greeting("Super Man") 

greeting(name="Super Man")

greet=greeting(name='Super Man')
print(greet)

print(greeting(name='Super Man'))

# all give the same
```
